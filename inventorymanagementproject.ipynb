```python
#Noting the time of tranction
import time

#Reading the Inventory
fd = open('data.txt','r')
products = fd.read().split('\n')
fd.close()

#Taking User Inputs
Username   = input("Enter your Name: ")
Usernumber = input("Enter your Number:")
Usermail   = input("Enter your E-mail:  ")
ui_prod_id = input("Enter The Product Id: ")
ui_prod_qn = input("Enter the Quantity: ")

updated_product_lst = []

# Going through each product details
for product in products:
    
    prod_details = product.split(',')
    
    if(prod_details[0] == ui_prod_id):
    #checking if product existing or not if yes
        if(int(ui_prod_qn) <= int(prod_details[3])):
        #We're having enough quantity 
            print('-'*40)
            print("product Name      :",prod_details[1])
            print("price             :",prod_details[2])
            print("Quantiy           :",ui_prod_qn)
            print('-'*40)
            bill = int(ui_prod_qn)*int(prod_details[2])
            print("Billing Amont     :",bill)
            print('-'*40)
            GST = bill * 0.18
            print("GST 18%             : " , GST)
            TotalAmount1 =  bill + (bill * 0.18)
            TransactionCharge = TotalAmount1 * 0.02
            print("transaction charges :",TransactionCharge  )
            TotalAmount = int(TotalAmount1 + TransactionCharge)
            print("Total Amount is     : ",   TotalAmount)
            
            #updating the inventory list
            prod_details[3] = str(int(prod_details[3]) - int(ui_prod_qn))
            
            # Generating sales in sales.txt
            fd = open("sales.txt",'a')
            sales_details = Username +","+ Usernumber +","+ Usermail +","+prod_details[1] +","+ui_prod_id +","+ui_prod_qn+","+str(int(ui_prod_qn) * int(prod_details[2])) +","+time.ctime()+'\n'
            fd.write(sales_details)
            fd.close()

        else:
            #When We are not enough quantity
            
            print("sorry,we're not having enough quantiy.")
            print("we're having only",prod_details[3],"quantity.")
            print("Would you like to purchase it?")
            
            ch = input("press Y/N: ")
            
            if (ch == 'y' or ch == 'Y'):
                # If you want to purchase with remaing quantity
                
                print('-'*40)
                print("product Name      :",prod_details[1])
                print("price             :",prod_details[2])
                print("Quantiy           :",prod_details[3])
                print('-'*40)
                bill = int(prod_details[3])*int(prod_details[2])
                print("Billing Amont     :",bill)
                print('-'*40)
                GST = bill * 0.18
                print("GST 18%             : " , GST)
                TotalAmount1 =  bill + (bill * 0.18)
                TransactionCharge = TotalAmount1 * 0.02
                print("transaction charges :",TransactionCharge  )
                TotalAmount = int(TotalAmount1 + TransactionCharge)
                print("Total Amount is     : ",   TotalAmount)
                
                # Generating sales in sales.txt
                fd = open("sales.txt",'a')
                sales_details = Username +","+ Usernumber +","+ Usermail +","+prod_details[1] +","+ui_prod_id +","+prod_details[3] +","+str(int(prod_details[3]) * int(prod_details[2])) +","+time.ctime()+'\n'
                fd.write(sales_details)
                fd.close()
                
                #updating Inventory list
                prod_details[3] = '0'
                
            #if checking product don't exist 
            else:
                print("thanks")

    #updating my inventory list
    updated_product_lst.append(prod_details)
    
lst =[]

#Updating my inventory String 
for i in updated_product_lst :
    prod = i[0] +","+ i[1] +","+ i[2] +","+ i[3] +'\n'
    lst.append(prod)
    
#removing last \n from the list
lst[-1] = lst[-1][:-1]
    

#Updating Inventory File
fd = open("data.txt",'w')

for i in lst:
    fd.write(i)
fd.close()
print('_' * 40)
print("Thank you ! Visit Again. ")
```
